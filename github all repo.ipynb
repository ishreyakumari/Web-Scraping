Fetch the all repositories of organisation ‘fossasia’ from Github. Print the name of repository.

```python
import json
import requests as r
i=1
python_data=[1]
while(len(python_data)!=0):
    parameters= {"page":i,"per_page":50}
    response = r.get("https://api.github.com/orgs/fossasia/repos",params=parameters)
    python_data = response.json()
    for j in python_data:
        print(j["name"])
    i+=1
```
